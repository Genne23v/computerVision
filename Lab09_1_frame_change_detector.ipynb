```python
import cv2 as cv
import time

cap = cv.VideoCapture(0)

if (cap.isOpened() == False):
    print("Error opening video stream or file")

frame_width = int(cap.get(3))
frame_height = int(cap.get(4))

# Save the first frame as background image.     
ret, background = cap.read()
background = cv.cvtColor(background, cv.COLOR_BGR2GRAY)
cv.imwrite('background.png', background)

i=0
while(cap.isOpened()):
    ret, frame = cap.read()
    if ret == False:
        break

    # Change code to display the difference images, instead of captured frames
    current_frame = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)
    diff = cv.absdiff(current_frame, background)
    # Threshold the difference image using a threshold of 128.
    thresh, binary = cv.threshold(diff, 128, 255, cv.THRESH_BINARY)
    # Count the number of pixels changed by at least 128 gray levels. 
    count = binary.sum()
    print(f'thres {thresh}, count {count}',)

    cv.imshow('Frame', binary)
    key = cv.waitKey(40)
        

    # If the count is larger than 1% of the number of pixels in the frame,
    if count > (frame_width * frame_height) * 0.05:
        i+=1
        current_time = time.time()
        # Output the time the activity was detected;
        print(f'Capture time: {current_time}')
        # Save the current color frame in a file (with a counter or time stamp)
        cv.imwrite(f'{current_time}.png', frame)
        # Wait for 5 seconds before returning to capturing loop. 
        cv.waitKey(5000)
        # Update the background image;
        background = current_frame
        cv.imwrite(f'diff_{i}.png', binary)
        cv.imwrite(f'background_{i}.png', background)
        
        
    if key & 0xFF == ord('q'):
        break

cap.release()
cv.destroyAllWindows()
```
